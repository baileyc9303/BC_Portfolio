the reason I chose the datasets I will be using is to find out what the villagers like, the gifts that everyone likes and which items a person should farm to use as gifts

```python
# First step is to import the url for the website
stardew_villagers_url = 'https://stardewvalleywiki.com/Villagers'
```

```python
# Second step I need to set a variable for requesting the HTML of the website.
stardew_villagers_html = requests.get(stardew_villagers_url).content

#Third step setting the variable for the HTML output using beautiful soup
stardew_villagers = BeautifulSoup(stardew_villagers_html,'html.parser')
```

```python
# At first I struggled to find exactly which part of the HTML all the villager's names were on, however, once I found it I used it to find all villager names.
#this will be used later to merge with other tables.
stardew_people = stardew_villagers.find_all('div', 'gallerytext')
```

```python
stardew_people
```

```python
#create a list of only the villager's names from the website.
character_names = [person.text.strip() for person in stardew_people]

stardew_names_df = pd.DataFrame(character_names, columns=['Character'])
stardew_names_df.head()
```

```python
# import character gift list
stardew_char_gifts_df = pd.read_csv('stardew_char_gifts.csv')
```

```python
# show the head of the set to remove unnecessary columns and rename the columns that are useful
stardew_char_gifts_df.head()
```

```python
# use the .drop method to remove the columns from the sheet that do not store any information
stardew_char_clean = stardew_char_gifts_df.drop(['Unnamed: 0', 'Unnamed: 6', 'Unnamed: 5'], axis=1)
stardew_char_clean.head()
```

```python
# rename columns that are titled "Unnamed" or the "1.07" column and replace with a name that shows what the info is showing
stardew_char_rename = stardew_char_clean.rename(columns={'Unnamed: 1': 'Character', 'Unnamed 2': 'Character Birthday', '1.07': 'Universal Likes', 'Unnamed: 7': 'Extras' })
```

```python
# remove the first row as it shows the names of the columns that were replaced.
stardew_char_likes_remove = stardew_char_rename.drop(stardew_char_rename.head(1).index)

# use .dropna to remove any unknown data in the set
stardew_char_likes = stardew_char_likes_remove.dropna(axis=1)
```

```python
stardew_char_likes
# Unnamed: 2 isn't changing names (fix for final dataset)
```

```python
# this step I grouped the gift and universal likes columns to find every item each villager likes
stardew_char_likes['All Gifts'] = stardew_char_likes.apply(lambda d: d['Gift Loves and Likes'] + d['Universal Likes'], axis=1).sum()
```

```python
stardew_char_likes
```

The dataset above was cleaned by removing data that was missing, as well as clearing out
the first row that was a repeat of the head of the dataset. Another step I took was renaming the heads of the
set because most of them were titled "Unnamed" from an error within the Google sheet. I also used the groupby
function to create a column titled "All Gifts" for every gift that each villager likes.

```python
# import stardew items url
stardew_items_url = 'https://raw.githubusercontent.com/Cecax27/DS-Stardew-Valley-Crops-Profit/refs/heads/main/data/crops_preprocessed_data.csv'
stardew_items_df = pd.read_csv(stardew_items_url)
```

```python
stardew_items_df.head()
```

```python
# this set has a lot of empty data so instead of removing columns or replacing the data
# i removed all the NaN items using .dropna
stardew_items_dropna = stardew_items_df.dropna(axis=1)
stardew_items_dropna.head()
```

```python
# take name and growth time columns from the stardew_items_dropna set for merging
stardew_items_cols = stardew_items_dropna[['name', 'growth_time']]
```

```python
# use pd.concat to merge both datasets and prepare for data cleaning
merged_stardew = pd.concat([stardew_items_cols, stardew_char_likes])
merged_stardew
```

```python
# use .fillna to change NaN to 'No Info'
stardew_full = merged_stardew.fillna('No Info')
stardew_full = stardew_full.rename(columns={'Unnamed: 2': 'Character Birthday'})
```
